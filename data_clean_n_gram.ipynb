```python
import re 
# from itertools import combinations
from collections import Counter
```

```python
file_names = ["dickens/chuzz.txt", "dickens/copperfield.txt", "dickens/donby.txt", "dickens/dorrit.txt", 
              "dickens/expectations.txt", "dickens/friend.txt", "dickens/house.txt", "dickens/nickleby.txt", 
              "dickens/pickwick.txt", "dickens/rudge.txt", "dickens/times.txt", "dickens/twist.txt"]
```

```python
name_indexes = ["chuzz", "copperfield", "donby", "dorrit", "expectations", "friend", "house", "nickleby", "pickwick", "rudge", "times", "twist"]
```

```python
name_pairs = zip(file_names, name_indexes)
```

```python
flattend_texts = {}

for pair in name_pairs:
    text_lines = []
    with open(pair[0]) as f:
        for line in f: 
            text_lines.append(line)

    
    cleaned_lines = []
    prev_was_newline = False

    for line in text_lines:
        if line == '\n':
            if not prev_was_newline:
                cleaned_lines.append('\n')
            prev_was_newline = True
        else:
            prev_was_newline = False

            line = line.rstrip('\n').lower()
            line = line.replace(' (', ", ").replace('(', ", ").replace(')', ",")
            line = line.replace(',,', ", ").replace('--', ", ")
            line = line.replace('_', '')
            line = re.sub(r'([^\w\s])', r' \1 ', line)
            line = re.sub(r'\s+', ' ', line).strip()

            cleaned_lines.append(line)
    

    text_array = []
    text_list = []
    for l in cleaned_lines: 
        if l == '\n':
            text_array.append(['\n'])
        else: 
            text_array.append(l.split(" "))

    for sect in text_array: 
        text_list.extend(sect)
    
    flattend_texts[pair[1]] = text_list
```

```python
for k in flattend_texts.keys():
    flat_text = flattend_texts[k]
    flat_text = [x for x in flat_text if x != 'ï']
    flat_text = [x for x in flat_text if x != '»']
    flat_text = [x for x in flat_text if x != '¿']

    flattend_texts[k] = flat_text
```

```python
total_words = []

for k in flattend_texts.keys():
    total_words.extend(flattend_texts[k])

total_words_set = set(total_words)

len(total_words_set)
```

```python
def ngram_counts(tokens, n):
    if n == 1:
        return Counter(tokens)
    return Counter(zip(*[tokens[i:] for i in range(n)]))
```

```python
unigram_count_per_book = []

for k in flattend_texts:
    text = flattend_texts[k]

    counts = Counter(text)  

    unigram_count_per_book.append(counts)
```

```python
total_unigram_counts = sum(unigram_count_per_book, Counter())
```

```python
bigram_count_per_book = []

for k in flattend_texts.keys():
    text = flattend_texts[k]

    counts = ngram_counts(text, 2)

    bigram_count_per_book.append(counts)

total_bigram_counts = sum(bigram_count_per_book, Counter())
```

```python
trigram_count_per_book = []

for k in flattend_texts.keys():
    text = flattend_texts[k]

    counts = ngram_counts(text, 3)

    trigram_count_per_book.append(counts)

total_trigram_counts = sum(trigram_count_per_book, Counter())
```

```python
fourgram_count_per_book = []

for k in flattend_texts.keys():
    text = flattend_texts[k]

    counts = ngram_counts(text, 4)

    fourgram_count_per_book.append(counts)

total_fourgram_counts = sum(fourgram_count_per_book, Counter())
```

```python
fivegram_count_per_book = []

for k in flattend_texts.keys():
    text = flattend_texts[k]

    counts = ngram_counts(text, 5)

    fivegram_count_per_book.append(counts)

total_fivegram_counts = sum(fivegram_count_per_book, Counter())
```

# Generating Probabilities and state matricies

```python
total_unigram_counts_dict = dict(total_unigram_counts)
total_unigram_probs = total_unigram_counts_dict.copy()

total_instances = 0
for key in total_unigram_counts_dict.keys():
    total_instances = total_unigram_counts_dict[key] + total_instances

for key in total_unigram_probs.keys():
    total_unigram_probs[key] = total_unigram_probs[key] / total_instances

total_unigram_probs
```

```python
total_bigram_counts_dict = dict(total_bigram_counts)
total_bigram_counts_dict
```

```python
bigram_states = {}

start_words = []
for pair in total_bigram_counts_dict.keys():
    start_words.append(pair[0])

start_words = set(start_words)

for word in start_words: 
    bigram_states[word] = {}

for pair in total_bigram_counts_dict.keys():
    bigram_states[pair[0]][pair[1]] = total_bigram_counts_dict[pair]


for start_wrd in bigram_states.keys():
    word_sum = 0 

    for second_word in bigram_states[start_wrd].keys():
        word_sum = word_sum + bigram_states[start_wrd][second_word]
    
    for second_word in bigram_states[start_wrd].keys():
        bigram_states[start_wrd][second_word] = bigram_states[start_wrd][second_word] / word_sum
```

```python
total_trigram_counts_dict = dict(total_trigram_counts)
total_trigram_counts_dict

trigram_states = {}

start_words_3 = []
for pair in total_trigram_counts_dict.keys():
    start_words_3.append((pair[0], pair[1]))

start_words_3 = set(start_words_3)
# start_words_t

for word in start_words_3: 
    trigram_states[word] = {}

for pair in total_trigram_counts_dict.keys():
    trigram_states[(pair[0], pair[1])][pair[2]] = total_trigram_counts_dict[pair]

# trigram_states
for start_wrd in trigram_states.keys():
    word_sum = 0 

    for second_word in trigram_states[start_wrd].keys():
        word_sum = word_sum + trigram_states[start_wrd][second_word]
    
    for second_word in trigram_states[start_wrd].keys():
        trigram_states[start_wrd][second_word] = trigram_states[start_wrd][second_word] / word_sum
```
