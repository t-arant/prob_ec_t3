```python
name_pairs = zip(file_names, name_indexes)
```

Data Cleaning

```python
flattend_texts = {}

for pair in name_pairs:
    text_lines = []
    with open(pair[0]) as f:
        for line in f:
            text_lines.append(line)


    cleaned_lines = []
    prev_was_newline = False

    for line in text_lines:
        if line == '\n':
            if not prev_was_newline:
                cleaned_lines.append('\n')
            prev_was_newline = True
        else:
            prev_was_newline = False

            line = line.rstrip('\n').lower()
            line = line.replace(' (', ", ").replace('(', ", ").replace(')', ",")
            line = line.replace(',,', ", ").replace('--', ", ")
            line = line.replace('_', '')
            line = re.sub(r'([^\w\s])', r' \1 ', line)
            line = re.sub(r'\s+', ' ', line).strip()

            cleaned_lines.append(line)


    text_array = []
    text_list = []
    for l in cleaned_lines:
        if l == '\n':
            text_array.append(['\n'])
        else:
            text_array.append(l.split(" "))

    for sect in text_array:
        text_list.extend(sect)

    flattend_texts[pair[1]] = text_list
```

```python
for k in flattend_texts.keys():
    flat_text = flattend_texts[k]
    flat_text = [x for x in flat_text if x != 'ï']
    flat_text = [x for x in flat_text if x != '»']
    flat_text = [x for x in flat_text if x != '¿']

    flattend_texts[k] = flat_text
```

N-Gram Counts

```python
def ngram_counts(tokens, n):
    if n == 1:
        return Counter(tokens)
    return Counter(zip(*[tokens[i:] for i in range(n)]))
```

```python
bigram_states = {}

start_words = []
for pair in total_bigram_counts_dict.keys():
    start_words.append(pair[0])

start_words = set(start_words)

for word in start_words:
    bigram_states[word] = {}

for pair in total_bigram_counts_dict.keys():
    bigram_states[pair[0]][pair[1]] = total_bigram_counts_dict[pair]


for start_wrd in bigram_states.keys():
    word_sum = 0

    for second_word in bigram_states[start_wrd].keys():
        word_sum = word_sum + bigram_states[start_wrd][second_word]

    for second_word in bigram_states[start_wrd].keys():
        bigram_states[start_wrd][second_word] = bigram_states[start_wrd][second_word] / word_sum
```

# **Using the Models to Make Predictions**

```python
import random

def _clean_sentence(sentence):
    line = sentence.lower()
    line = line.replace(' (', ", ").replace('(', ", ").replace(')', ",")
    line = line.replace(',,', ", ").replace('--', ", ")
    line = line.replace('_', '')
    line = re.sub(r'([^\w\s])', r' \1 ', line)
    line = re.sub(r'\s+', ' ', line).strip()
    return line.split(" ")

def predict_next_word_bigram(sentence):
    cleaned_tokens = _clean_sentence(sentence)
    last_word = cleaned_tokens[-1]

    if last_word in bigram_states:
        possible_next_words = list(bigram_states[last_word].keys())
        probabilities = list(bigram_states[last_word].values())
        if possible_next_words:
            return random.choices(possible_next_words, weights=probabilities, k=1)[0]
    return "[UNKNOWN]"

def predict_next_word_trigram(sentence):
    cleaned_tokens = _clean_sentence(sentence)
    if len(cleaned_tokens) < 2:
        return predict_next_word_bigram(sentence) # Fallback to bigram if not enough words

    last_two_words = tuple(cleaned_tokens[-2:])

    if last_two_words in trigram_states:
        possible_next_words = list(trigram_states[last_two_words].keys())
        probabilities = list(trigram_states[last_two_words].values())
        if possible_next_words:
            return random.choices(possible_next_words, weights=probabilities, k=1)[0]
    return predict_next_word_bigram(sentence) # Fallback to bigram if trigram not found

def predict_next_word_fourgram(sentence):
    cleaned_tokens = _clean_sentence(sentence)
    if len(cleaned_tokens) < 3:
        return predict_next_word_trigram(sentence) # Fallback to trigram if not enough words

    last_three_words = tuple(cleaned_tokens[-3:])

    if last_three_words in fourgram_states:
        possible_next_words = list(fourgram_states[last_three_words].keys())
        probabilities = list(fourgram_states[last_three_words].values())
        if possible_next_words:
            return random.choices(possible_next_words, weights=probabilities, k=1)[0]
    return predict_next_word_trigram(sentence) # Fallback to trigram if fourgram not found

def predict_next_word_fivegram(sentence):
    cleaned_tokens = _clean_sentence(sentence)
    if len(cleaned_tokens) < 4:
        return predict_next_word_fourgram(sentence)  # Fallback to fourgram if not enough words

    last_four_words = tuple(cleaned_tokens[-4:])
    if last_four_words in fivegram_states:
        possible_next_words = list(fivegram_states[last_four_words].keys())
        probabilities = list(fivegram_states[last_four_words].values())
        if possible_next_words:
            return random.choices(possible_next_words, weights=probabilities, k=1)[0]
    return predict_next_word_fourgram(sentence)  # Fallback to fourgram if fivegram not found
```

# **Autoregressive Generation**

The next section takes a phrase and autoregressively completes it for n number of times. The input here is the same phrase as above.

```python
def generate_full_sentence(model_type, initial_sentence, num_words_to_generate):
    current_sentence = initial_sentence
    generated_words = []

    for _ in range(num_words_to_generate):
        next_word = ""
        if model_type == "bigram":
            next_word = predict_next_word_bigram(current_sentence)
        elif model_type == "trigram":
            next_word = predict_next_word_trigram(current_sentence)
        elif model_type == "fourgram":
            next_word = predict_next_word_fourgram(current_sentence)
        elif model_type == "fivegram":
            next_word = predict_next_word_fivegram(current_sentence)
        else:
            return "Invalid model type specified."

        if next_word == "[UNKNOWN]":
            # If a model can't predict, try a simpler model or stop
            if model_type == "fivegram":
                next_word = predict_next_word_fourgram(current_sentence) if predict_next_word_fourgram(current_sentence) != "[UNKNOWN]" else predict_next_word_trigram(current_sentence)
            elif model_type == "fourgram":
                next_word = predict_next_word_trigram(current_sentence) if predict_next_word_trigram(current_sentence) != "[UNKNOWN]" else predict_next_word_bigram(current_sentence)
            elif model_type == "trigram":
                next_word = predict_next_word_bigram(current_sentence)

            if next_word == "[UNKNOWN]":
                break # Stop if even bigram can't predict

        current_sentence += " " + next_word
        generated_words.append(next_word)

    return ' '.join([word for word in _clean_sentence(initial_sentence) if word not in [',', '.', '!', '?', ';', ':']] + generated_words)

```

# **Evaluation**

We evaluate each model on the next word prediction task. Accuracy is given by (Number of Correct Predictions)/(Total Number of Words in Task)

```python
def evaluate_model(model_type, book_name, prediction_limit):
    word_list = flattend_texts[book_name]
    correct_predictions = 0
    total_predictions = 0

    # Determine starting index based on model_type to ensure enough context
    start_index = 0
    if model_type == "bigram":
        start_index = 1  # Needs 1 preceding word (index i-1)
    elif model_type == "trigram":
        start_index = 2  # Needs 2 preceding words (indices i-2, i-1)
    elif model_type == "fourgram":
        start_index = 3  # Needs 3 preceding words (indices i-3, i-2, i-1)
    elif model_type == "fivegram":
        start_index = 4  # Needs 4 preceding words (indices i-4, i-3, i-2, i-1)
    else:
        print("Invalid model type.")
        return 0, 0

    for i in range(start_index, len(word_list)):
        if total_predictions >= prediction_limit:
            break

        context_words = []
        if model_type == "bigram":
            context_words = [word_list[i-1]]
        elif model_type == "trigram":
            context_words = [word_list[i-2], word_list[i-1]]
        elif model_type == "fourgram":
            context_words = [word_list[i-3], word_list[i-2], word_list[i-1]]
        elif model_type == "fivegram":
            context_words = [word_list[i-4], word_list[i-3], word_list[i-2], word_list[i-1]]

        context_sentence = " ".join(context_words)
        target_word = word_list[i]

        predicted_word = "[UNKNOWN]"
        if model_type == "bigram":
            predicted_word = predict_next_word_bigram(context_sentence)
        elif model_type == "trigram":
            predicted_word = predict_next_word_trigram(context_sentence)
        elif model_type == "fourgram":
            predicted_word = predict_next_word_fourgram(context_sentence)
        elif model_type == "fivegram":
            predicted_word = predict_next_word_fivegram(context_sentence)

        if predicted_word == target_word:
            correct_predictions += 1

        total_predictions += 1

    return correct_predictions/total_predictions

```

```python
def evaluate_total_books(model_type, prediction_limit):
    # Use the global total_words list
    word_list = total_words
    correct_predictions = 0
    total_predictions = 0

    # Determine starting index based on model_type to ensure enough context
    start_index = 0
    if model_type == "bigram":
        start_index = 1  # Needs 1 preceding word (index i-1)
    elif model_type == "trigram":
        start_index = 2  # Needs 2 preceding words (indices i-2, i-1)
    elif model_type == "fourgram":
        start_index = 3  # Needs 3 preceding words (indices i-3, i-2, i-1)
    elif model_type == "fivegram":
        start_index = 4  # Needs 4 preceding words (indices i-4, i-3, i-2, i-1)
    else:
        print("Invalid model type.")
        return 0

    for i in range(start_index, len(word_list)):
        if total_predictions >= prediction_limit:
            break

        context_words = []
        if model_type == "bigram":
            context_words = [word_list[i-1]]
        elif model_type == "trigram":
            context_words = [word_list[i-2], word_list[i-1]]
        elif model_type == "fourgram":
            context_words = [word_list[i-3], word_list[i-2], word_list[i-1]]
        elif model_type == "fivegram":
            context_words = [word_list[i-4], word_list[i-3], word_list[i-2], word_list[i-1]]

        context_sentence = " ".join(context_words)
        target_word = word_list[i]

        predicted_word = "[UNKNOWN]"
        if model_type == "bigram":
            predicted_word = predict_next_word_bigram(context_sentence)
        elif model_type == "trigram":
            predicted_word = predict_next_word_trigram(context_sentence)
        elif model_type == "fourgram":
            predicted_word = predict_next_word_fourgram(context_sentence)
        elif model_type == "fivegram":
            predicted_word = predict_next_word_fivegram(context_sentence)

        if predicted_word == target_word:
            correct_predictions += 1

        total_predictions += 1

    if total_predictions == 0:
        return 0
    return correct_predictions/total_predictions
```
